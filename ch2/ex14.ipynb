# Extension questions
1. Set all the odd temperatures to the mean of all the temperatures.
2. Set the even temperatures at hours 9 and 18 to 3.
3. If the hour is odd, set the temperature to 5.

```python
temps = pd.read_csv("../data/nyc-temps.txt", header=None).squeeze().astype(np.float64)
df = pd.DataFrame(
    {"temp": temps, "hour": [0, 3, 6, 9, 12, 15, 18, 21] * (temps.count() // 8) + [0]}
)

# 1. All odd tempeatures to the mean of all temperatures
df.loc[df["temp"] % 2 == 1, "temp"] = df["temp"].mean()
df.describe()
```

```python
temps = pd.read_csv("../data/nyc-temps.txt", header=None).squeeze().astype(np.float64)
df = pd.DataFrame(
    {"temp": temps, "hour": [0, 3, 6, 9, 12, 15, 18, 21] * (temps.count() // 8) + [0]}
)
# 2. Set the even temperatures at hours 9 and 18 to 3
df.loc[(df["temp"] % 2 == 0) & df["hour"].isin([9, 18]), "temp"] = 3
df.describe()
```

```python
temps = pd.read_csv("../data/nyc-temps.txt", header=None).squeeze().astype(np.float64)
df = pd.DataFrame(
    {"temp": temps, "hour": [0, 3, 6, 9, 12, 15, 18, 21] * (temps.count() // 8) + [0]}
)

# 3. If the hour is odd, set the temperature to 5
df.loc[df["hour"] % 2 == 1, "temp"] = 5
df.describe()
```
